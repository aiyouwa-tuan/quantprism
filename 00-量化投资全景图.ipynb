## 一、量化投资到底是什么？

一句话：**用数据和规则代替直觉做投资决策**。

| | 主观投资（手动交易） | 量化投资 |
|---|---|---|
| 决策依据 | 感觉、经验、消息 | 数据、模型、规则 |
| 执行方式 | 手动下单 | 程序自动执行 |
| 情绪影响 | 大：恐惧、贪婪、犹豫 | 小：按规则执行 |
| 覆盖范围 | 盯几只股票 | 同时监控几百上千只 |
| 回测验证 | 靠记忆，模糊 | 精确回测，看历史表现 |
| 典型问题 | "我觉得该涨了" | "历史上这种形态73%会涨" |

### 一个简单例子

**主观交易者的一天：**
> 早上看新闻说 BTC 要涨 → 买了 → 下午跌了3% → 慌了 → 割肉 → 晚上又涨回来了 → 后悔

**量化交易者的一天：**
> 策略规则：RSI < 30 且成交量放大 2 倍 → 买入，止损 -5%，止盈 +10% → 程序自动执行 → 去喝咖啡

量化不是魔法，不保证赚钱。它的核心价值是：**把你的投资逻辑变成可验证、可重复、不受情绪干扰的系统**。

## 二、量化投资的完整链条

量化投资不是「写个程序自动买卖」这么简单，它是一条完整的生产线：

```
┌─────────────────────────────────────────────────────────────────────┐
│                     量化投资的 7 个环节                              │
├─────────────────────────────────────────────────────────────────────┤
│                                                                     │
│  ① 市场认知    →  你对市场的理解和假设                                │
│       ↓           "BTC在恐慌时往往超跌后反弹"                        │
│                                                                     │
│  ② 策略设计    →  把认知变成可执行的规则                              │
│       ↓           "RSI<25 + 恐慌指数>80 → 买入"                     │
│                                                                     │
│  ③ 数据获取    →  获取验证策略需要的数据                              │
│       ↓           历史价格、成交量、链上数据、新闻情绪...              │
│                                                                     │
│  ④ 回测验证    →  用历史数据检验策略是否有效                          │
│       ↓           "过去5年，这策略年化收益38%，最大回撤-22%"           │
│                                                                     │
│  ⑤ 风险管理    →  控制每笔交易和总仓位的风险                          │
│       ↓           "单笔最多亏总资金的2%，总仓位不超过60%"             │
│                                                                     │
│  ⑥ 实盘执行    →  先模拟盘验证，再小资金实盘                          │
│       ↓           程序自动下单，处理网络异常、滑点等                    │
│                                                                     │
│  ⑦ 监控迭代    →  持续监控表现，策略失效时调整                        │
│                   "最近3个月夏普比率从1.8降到0.3，需要排查"            │
│                                                                     │
└─────────────────────────────────────────────────────────────────────┘
```

## 三、量化策略的四大流派

所有量化策略，归根结底都在赌一个观点：

| 流派 | 核心逻辑 | 赌什么 | 典型策略 | 适合市场 |
|------|---------|--------|---------|----------|
| **趋势跟踪** | 涨的会继续涨，跌的会继续跌 | 趋势会延续 | 均线交叉、突破 | 趋势明显的市场（加密货币） |
| **均值回归** | 涨多了会跌回来，跌多了会涨回来 | 价格会回到正常 | RSI超买超卖、布林带 | 震荡市、成熟市场（美股蓝筹）|
| **统计套利** | 相似的东西价格应该差不多 | 价差会收敛 | 配对交易、跨交易所搬砖 | 多品种市场 |
| **事件驱动** | 特定事件后价格有规律性反应 | 历史会重演 | 财报后交易、减半行情 | 有定期事件的市场 |

### 四个流派的形象比喻

```
🏄 趋势跟踪 = 冲浪
   浪来了就跟着冲，浪没了就上岸。
   不预测浪什么时候来，只在浪出现时跟上。
   ✅ 赚大钱靠大趋势  ❌ 横盘时反复被打脸

🪀 均值回归 = 橡皮筋
   价格像橡皮筋，拉得越远，弹回来的力量越大。
   在极端位置反向操作。
   ✅ 胜率高，适合稳健型  ❌ 遇到真趋势会亏惨（橡皮筋断了）

⚖️ 统计套利 = 找不平衡
   A交易所BTC $65000，B交易所 $65200 → 低买高卖赚差价。
   或者：可口可乐和百事可乐股价通常同涨同跌，某天分道扬镳 → 赌它们回归。
   ✅ 风险低，收益稳  ❌ 机会少，竞争激烈

📅 事件驱动 = 看日历赚钱
   BTC减半前3个月通常上涨 → 提前布局。
   苹果发布财报后如果超预期 → 追涨。
   ✅ 逻辑清晰，好理解  ❌ 样本少，不一定重演
```

## 四、决定赚亏的 5 个核心因素

### 这是整个文档最重要的部分。

````python
import numpy as np

# 模拟：同一个策略，不同风控下的结果差异
np.random.seed(42)
n_trades = 200
win_rate = 0.45  # 胜率45%（不高）
win_size = 3.0   # 赢的时候平均赚3%
lose_size = 2.0  # 输的时候平均亏2%

# 策略A：不控风险，每次满仓
equity_a = [100000]
for i in range(n_trades):
    if np.random.random() < win_rate:
        equity_a.append(equity_a[-1] * (1 + win_size/100 * 5))  # 满仓，赢赚15%
    else:
        equity_a.append(equity_a[-1] * (1 - lose_size/100 * 5))  # 满仓，输亏10%

# 策略B：严格风控，每次只用20%仓位
equity_b = [100000]
for i in range(n_trades):
    if np.random.random() < win_rate:
        equity_b.append(equity_b[-1] * (1 + win_size/100 * 1))  # 20%仓位，赢赚3%
    else:
        equity_b.append(equity_b[-1] * (1 - lose_size/100 * 1))  # 20%仓位，输亏2%

# 策略C：有风控 + 盈亏比优化（让利润奔跑，截断亏损）
np.random.seed(42)
equity_c = [100000]
for i in range(n_trades):
    if np.random.random() < win_rate:
        # 赢了加仓，让利润奔跑
        equity_c.append(equity_c[-1] * (1 + win_size/100 * 1.8))
    else:
        # 输了快速止损
        equity_c.append(equity_c[-1] * (1 - lose_size/100 * 0.5))

fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(equity_a, label='A: 满仓梭哈（无风控）', color='red', linewidth=1.5, alpha=0.8)
ax.plot(equity_b, label='B: 固定20%仓位', color='steelblue', linewidth=1.5)
ax.plot(equity_c, label='C: 风控+让利润奔跑', color='green', linewidth=2)
ax.axhline(y=100000, color='gray', linestyle='--', linewidth=0.5)
ax.set_title('同一个策略（胜率45%），不同风控方式的结果天差地别', fontsize=14, fontweight='bold')
ax.set_xlabel('交易次数')
ax.set_ylabel('资金 ($)')
ax.legend(fontsize=12)
ax.set_yscale('log')  # 对数坐标更直观
ax.grid(True, alpha=0.3)
ax.text(150, equity_a[-1], f'${equity_a[-1]:,.0f}', color='red', fontsize=11)
ax.text(150, equity_b[-1], f'${equity_b[-1]:,.0f}', color='steelblue', fontsize=11)
ax.text(150, equity_c[-1], f'${equity_c[-1]:,.0f}', color='green', fontsize=11)
plt.tight_layout()
plt.show()

print("三种方式的最终资金：")
print(f"  A 满仓梭哈:    ${equity_a[-1]:>12,.0f}  ({'赚' if equity_a[-1]>100000 else '亏'})")
print(f"  B 固定仓位:    ${equity_b[-1]:>12,.0f}  ({'赚' if equity_b[-1]>100000 else '亏'})")
print(f"  C 风控+止盈:   ${equity_c[-1]:>12,.0f}  ({'赚' if equity_c[-1]>100000 else '亏'})")
print(f"\n三个策略的胜率完全一样（45%），差别只在资金管理！")
````

### 因素1：盈亏比 > 胜率

**很多人只关注「准不准」（胜率），但真正赚钱的关键是「赚多少 vs 亏多少」（盈亏比）。**

| | 胜率 | 平均盈利 | 平均亏损 | 盈亏比 | 100次后结果 |
|---|:---:|:---:|:---:|:---:|---|
| 策略A | 70% | +1% | -3% | 0.33 | 70×1% - 30×3% = **-20%** 亏！ |
| 策略B | 40% | +5% | -2% | 2.5 | 40×5% - 60×2% = **+80%** 赚！ |

**策略A胜率高达70%但亏钱，策略B胜率只有40%但赚大钱。**

**核心公式：期望收益 = 胜率 × 平均盈利 - (1-胜率) × 平均亏损**

> 这个数字 > 0 才值得做。等于看你每做一次交易，**平均**能赚多少。

````python
# 可视化：胜率 vs 盈亏比 的盈利地图
win_rates = np.arange(0.2, 0.85, 0.05)
ratios = np.arange(0.5, 5.1, 0.25)  # 盈亏比

W, R = np.meshgrid(win_rates, ratios)
# 期望收益 = 胜率 * 盈亏比 - (1-胜率)
expected = W * R - (1 - W)

fig, ax = plt.subplots(figsize=(12, 8))
c = ax.contourf(W*100, R, expected, levels=20, cmap='RdYlGn', alpha=0.9)
ax.contour(W*100, R, expected, levels=[0], colors='black', linewidths=2)  # 盈亏平衡线
plt.colorbar(c, label='每笔交易期望收益（倍）')

ax.set_xlabel('胜率 (%)', fontsize=13)
ax.set_ylabel('盈亏比（赚/亏）', fontsize=13)
ax.set_title('盈利地图：绿色区域赚钱，红色区域亏钱\n黑线 = 盈亏平衡线', fontsize=14, fontweight='bold')

# 标注几个典型位置
ax.plot(70, 0.33, 'rx', markersize=15, markeredgewidth=3)
ax.annotate('策略A\n胜率高但亏', xy=(70, 0.33), xytext=(75, 1), fontsize=10,
           arrowprops=dict(arrowstyle='->', color='red'), color='red')

ax.plot(40, 2.5, 'g*', markersize=20)
ax.annotate('策略B\n胜率低但赚', xy=(40, 2.5), xytext=(28, 3.5), fontsize=10,
           arrowprops=dict(arrowstyle='->', color='green'), color='green')

ax.plot(50, 2.0, 'b^', markersize=15)
ax.annotate('理想区域\n50%胜率+2倍盈亏比', xy=(50, 2.0), xytext=(55, 3.5), fontsize=10,
           arrowprops=dict(arrowstyle='->', color='blue'), color='blue')

plt.tight_layout()
plt.show()
````

### 因素2：风险管理（仓位控制）

上面的模拟已经说明了一切。这里给出具体原则：

```
🔴 致命错误（90%的新手亏在这里）
├── 满仓操作：一把亏完没有翻身机会
├── 不设止损："再等等" → 亏10% → 亏30% → 亏60%
├── 亏了加仓："摊低成本" → 越亏越多
└── 频繁交易：手续费吃掉利润

🟢 正确做法
├── 单笔风险 ≤ 总资金的1-2%（亏了不心疼，不影响判断）
├── 总仓位 ≤ 50-70%（留现金应对黑天鹅）
├── 每笔交易必有止损（进场前就定好）
└── 盈亏比 ≥ 2:1（赚的时候至少是亏的2倍）
```

**为什么 1-2% 这么重要？看数学：**

| 单笔风险 | 连亏10次后剩余 | 连亏20次后剩余 | 心理状态 |
|:---:|:---:|:---:|---|
| 1% | 90.4% | 81.8% | 很淡定，正常波动 |
| 2% | 81.7% | 66.8% | 能接受，继续执行 |
| 5% | 59.9% | 35.8% | 开始怀疑策略 |
| 10% | 34.9% | 12.2% | 崩溃，开始乱操作 |
| 20% | 10.7% | 1.2% | 已经爆仓了 |

### 因素3：策略是否有「真实边际」（Alpha）

**边际（Edge）= 你比市场其他参与者多知道的东西，或做得更好的地方。**

没有边际的策略 = 掷硬币 + 手续费 = **必亏**。

| 边际来源 | 举例 | 持续性 |
|---------|------|--------|
| **信息差** | 你比别人更早知道某个消息 | 低（很快被消化）|
| **分析差** | 同样的数据你解读得更准确 | 中（可学习）|
| **行为差** | 别人恐慌抛售时你按规则买入 | 高（人性不变）|
| **速度差** | 你的程序比别人快 | 低（军备竞赛）|
| **耐心差** | 别人拿不住，你能拿住 | 高 |

**对散户最现实的边际是「行为差」和「耐心差」：**
> 用量化系统克服恐惧和贪婪，在别人情绪崩溃时严格执行策略。
> 这就是为什么量化对散户有价值 —— 不是因为策略多厉害，而是它让你不犯低级错误。

### 因素4：过拟合 —— 量化投资的头号杀手

````python
# 演示过拟合：用同一组数据反复优化，看起来很好但实际没用
np.random.seed(42)

# 模拟「真实市场」：其实是随机的，没有任何规律
n_days = 500
random_prices = 100 * np.exp(np.cumsum(np.random.randn(n_days) * 0.02))

# 「过拟合策略」：找历史上最完美的参数组合
# 通过大量尝试，总能找到一组参数在历史上看起来很棒
best_return = -999
best_params = None

for short_ma in range(3, 30):
    for long_ma in range(short_ma+5, 60):
        # 简单均线交叉
        short = np.convolve(random_prices, np.ones(short_ma)/short_ma, mode='valid')
        long_v = np.convolve(random_prices, np.ones(long_ma)/long_ma, mode='valid')
        min_len = min(len(short), len(long_v))
        short = short[-min_len:]
        long_v = long_v[-min_len:]
        prices_trim = random_prices[-min_len:]
        
        signal = np.where(short > long_v, 1, -1)
        returns = np.diff(prices_trim) / prices_trim[:-1] * signal[:-1]
        total = np.sum(returns) * 100
        
        if total > best_return:
            best_return = total
            best_params = (short_ma, long_ma)

# 看看这个「最优策略」在新数据上表现如何
np.random.seed(99)  # 不同的随机种子 = 未来的数据
future_prices = 100 * np.exp(np.cumsum(np.random.randn(n_days) * 0.02))

sm, lm = best_params
short = np.convolve(future_prices, np.ones(sm)/sm, mode='valid')
long_v = np.convolve(future_prices, np.ones(lm)/lm, mode='valid')
min_len = min(len(short), len(long_v))
signal = np.where(short[-min_len:] > long_v[-min_len:], 1, -1)
prices_trim = future_prices[-min_len:]
returns_future = np.diff(prices_trim) / prices_trim[:-1] * signal[:-1]
future_return = np.sum(returns_future) * 100

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(['历史回测'], [best_return], color='green', width=0.5)
axes[0].set_title(f'回测结果（参数: MA{sm}/{lm}）', fontsize=13)
axes[0].set_ylabel('总收益率 (%)')
axes[0].text(0, best_return + 2, f'+{best_return:.1f}%', ha='center', fontsize=14, fontweight='bold', color='green')

c = 'red' if future_return < 0 else 'green'
axes[1].bar(['实盘表现'], [future_return], color=c, width=0.5)
axes[1].set_title('同一策略在新数据上的表现', fontsize=13)
axes[1].set_ylabel('总收益率 (%)')
axes[1].text(0, future_return + (2 if future_return > 0 else -5), f'{future_return:+.1f}%', 
             ha='center', fontsize=14, fontweight='bold', color=c)
axes[1].axhline(y=0, color='gray', linestyle='--')

fig.suptitle('过拟合陷阱：在随机数据上也能"找到"漂亮的策略', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f"这组数据其实是完全随机的（没有任何规律）！")
print(f"但通过测试 {27*30}+ 种参数组合，我们'发现'了一个回测收益 +{best_return:.1f}% 的策略")
print(f"拿到新数据上一跑：{future_return:+.1f}% → 原形毕露")
print(f"\n这就是过拟合：你以为发现了规律，其实只是在噪音中找到了巧合。")
````

**如何避免过拟合？**

```
✅ 留出样本外数据：用 2018-2022 年数据开发策略，用 2023-2024 检验
✅ 参数越少越好：2-3 个参数的策略比 10 个参数的可靠得多
✅ 逻辑要说得通：不能只是「历史上这么做赚钱」，要有合理解释
✅ 多市场验证：在 BTC 上有效，在 ETH、SOL 上也有效 → 更可信
❌ 不要反复优化参数直到回测好看
❌ 不要用太短的数据（至少要覆盖一个牛熊周期）
```

### 因素5：交易成本与滑点

**很多回测中忽略的「隐形杀手」：**

````python
# 交易成本对策略的侵蚀
trades_per_year = [10, 50, 100, 250, 500, 1000, 2500]
cost_per_trade = 0.1  # 单次交易成本 0.1%（手续费+滑点）

gross_return = 20  # 策略毛收益 20%
net_returns = [gross_return - n * cost_per_trade for n in trades_per_year]

fig, ax = plt.subplots(figsize=(12, 6))
colors = ['green' if r > 0 else 'red' for r in net_returns]
bars = ax.bar([str(n) for n in trades_per_year], net_returns, color=colors, edgecolor='white', width=0.6)
ax.axhline(y=0, color='black', linewidth=1)
ax.axhline(y=20, color='green', linewidth=1, linestyle='--', alpha=0.5, label='毛收益 20%')

for bar, val, n in zip(bars, net_returns, trades_per_year):
    cost = n * cost_per_trade
    ax.text(bar.get_x() + bar.get_width()/2, val + 1, 
            f'{val:+.0f}%\n(成本{cost:.0f}%)', ha='center', fontsize=10)

ax.set_xlabel('每年交易次数', fontsize=12)
ax.set_ylabel('净收益率 (%)', fontsize=12)
ax.set_title('交易频率 vs 净收益（假设毛收益20%，每次成本0.1%）', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("结论：")
print("  交易越频繁，成本越高。年交易 200+ 次时，20%的毛收益就被吃掉大半。")
print("  对散户来说，中低频策略（月/周级别）更现实。")
print("\n各市场典型成本：")
print("  加密货币: 0.04-0.1% 每笔（Maker/Taker）")
print("  美股:     $0（零佣金，但有 spread 约 0.01-0.05%）")
````

## 五、赚钱策略 vs 亏钱策略的特征对比

````python
# 模拟三种策略的资金曲线
np.random.seed(42)
days = 500

# 好策略：稳定向上，回撤小
good = [100000]
for _ in range(days):
    good.append(good[-1] * (1 + np.random.normal(0.0005, 0.008)))

# 坏策略：大起大落，最终亏损
np.random.seed(7)
bad = [100000]
for _ in range(days):
    bad.append(bad[-1] * (1 + np.random.normal(-0.0002, 0.025)))

# 过拟合策略：一开始好看，后面崩盘
np.random.seed(42)
overfit = [100000]
for i in range(days):
    if i < 300:  # 前300天=回测期
        overfit.append(overfit[-1] * (1 + np.random.normal(0.001, 0.005)))
    else:  # 后200天=实盘
        overfit.append(overfit[-1] * (1 + np.random.normal(-0.001, 0.015)))

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# 好策略
axes[0].plot(good, color='green', linewidth=1.5)
axes[0].axhline(y=100000, color='gray', linestyle='--', linewidth=0.5)
axes[0].set_title('好策略\n稳定上升，回撤可控', fontsize=12, color='green')
axes[0].set_ylabel('资金')
ret = (good[-1]/good[0]-1)*100
axes[0].text(250, max(good)*0.95, f'收益: +{ret:.1f}%', fontsize=12, color='green')

# 坏策略  
axes[1].plot(bad, color='red', linewidth=1.5)
axes[1].axhline(y=100000, color='gray', linestyle='--', linewidth=0.5)
axes[1].set_title('坏策略\n大起大落，趋势向下', fontsize=12, color='red')
ret2 = (bad[-1]/bad[0]-1)*100
axes[1].text(250, max(bad)*0.95, f'收益: {ret2:+.1f}%', fontsize=12, color='red')

# 过拟合
axes[2].plot(overfit, color='orange', linewidth=1.5)
axes[2].axhline(y=100000, color='gray', linestyle='--', linewidth=0.5)
axes[2].axvline(x=300, color='red', linestyle='--', linewidth=1)
axes[2].set_title('过拟合策略\n回测漂亮，实盘崩溃', fontsize=12, color='orange')
axes[2].text(100, max(overfit)*0.85, '← 回测期 →', fontsize=10, color='gray')
axes[2].text(350, max(overfit)*0.85, '实盘', fontsize=10, color='red')

for ax in axes:
    ax.grid(True, alpha=0.3)
    ax.set_xlabel('天数')

plt.tight_layout()
plt.show()
````

### 一眼判断策略好坏的指标

| 指标 | 含义 | 好策略 | 差策略 |
|------|------|:------:|:------:|
| **夏普比率** | 每承担1份风险赚多少 | > 1.5 | < 0.5 |
| **最大回撤** | 最惨的时候亏多少 | < -20% | > -50% |
| **盈亏比** | 平均赚 / 平均亏 | > 2 | < 1 |
| **胜率** | 赚钱交易的比例 | > 40% | < 30% |
| **年化收益** | 每年赚多少 | > 15% | < 无风险利率 |
| **卡尔玛比率** | 年化收益 / 最大回撤 | > 1 | < 0.5 |
| **交易次数** | 样本是否够 | > 100 | < 30 |

**最重要的三个：夏普比率、最大回撤、盈亏比。**

通俗解释：
- **夏普比率**：你的收益是靠实力还是靠运气？越高越靠实力。
- **最大回撤**：最坏情况你能不能扛住？这决定你会不会半路放弃。
- **盈亏比**：你赚一次能不能覆盖亏好几次？
