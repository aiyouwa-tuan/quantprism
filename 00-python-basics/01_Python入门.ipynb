## 2️⃣ 变量 — 给数据取个名字

变量就像一个**盒子**，你把数据放进去，之后用名字取出来。

```python
# 📊 股票代码（字符串）
stock_code = "AAPL"          # 苹果公司
crypto_symbol = "BTC/USDT"   # 比特币

# 💰 价格（浮点数）
stock_price = 178.52
btc_price = 65432.10

# 📦 持仓数量（整数）
shares = 100
btc_amount = 2

# ✅ 是否持仓（布尔值）
is_holding = True

# 打印看看
print(f"股票: {stock_code}  价格: ${stock_price}  持有: {shares}股")
print(f"加密: {crypto_symbol}  价格: ${btc_price:,}  持有: {btc_amount}个")
```

### 🔍 查看数据类型
`type()` 函数告诉你变量是什么类型：

```python
print(f"stock_code → {type(stock_code).__name__}（字符串）")
print(f"stock_price → {type(stock_price).__name__}（浮点数）")
print(f"shares      → {type(shares).__name__}（整数）")
print(f"is_holding  → {type(is_holding).__name__}（布尔值）")
```

## 3️⃣ 数学运算 — 计算你的持仓市值

```python
# 计算持仓市值
stock_value = stock_price * shares
btc_value = btc_price * btc_amount
total = stock_value + btc_value

print("┌──────────────────────────────────┐")
print("│        💼 我的持仓明细            │")
print("├──────────────────────────────────┤")
print(f"│  AAPL   {shares}股 × ${stock_price}  = ${stock_value:>10,.2f} │")
print(f"│  BTC    {btc_amount}个  × ${btc_price:,} = ${btc_value:>10,.2f} │")
print("├──────────────────────────────────┤")
print(f"│  💰 总资产               ${total:>10,.2f} │")
print("└──────────────────────────────────┘")
```

## 4️⃣ 实战：计算收益率 📈

**收益率公式**：

$$收益率 = \frac{卖出价 - 买入价}{买入价} \times 100\%$$

```python
buy_price = 150.00    # 买入价
sell_price = 178.52   # 当前价
shares = 100          # 持有数量

profit = (sell_price - buy_price) * shares
profit_rate = (sell_price - buy_price) / buy_price * 100

print(f"买入: ${buy_price}  →  卖出: ${sell_price}")
print(f"盈亏: ${profit:,.2f}")
print(f"收益率: {profit_rate:.2f}%")

if profit > 0:
    print("\n🎉 恭喜赚钱了！")
else:
    print("\n😢 亏了，需要复盘策略")
```

## 5️⃣ 字符串操作 — 处理股票代码

```python
symbol = "  aapl  "

# 去空格 + 转大写（从 API 拿到的数据经常需要清洗）
clean_symbol = symbol.strip().upper()

print(f"原始: '{symbol}'")
print(f"清洗后: '{clean_symbol}'")

# 拼接完整代码
full_code = "NASDAQ:" + clean_symbol
print(f"完整代码: {full_code}")
```

## 6️⃣ 类型转换 — API 返回的数据经常是字符串

```python
# 从 API 拿到的数据通常是字符串，需要转换
price_str = "178.52"
volume_str = "1000000"

price = float(price_str)    # 字符串 → 浮点数
volume = int(volume_str)    # 字符串 → 整数

print(f"价格: {price} (类型: {type(price).__name__})")
print(f"成交量: {volume:,} (类型: {type(volume).__name__})")

# 现在可以做数学运算了
market_value = price * volume
print(f"成交额: ${market_value:,.2f}")
```
