## 1️⃣ 比较运算符

| 运算符 | 含义 | 例子 |
|:---:|------|------|
| `>` | 大于 | `price > 100` |
| `<` | 小于 | `rsi < 30` |
| `>=` | 大于等于 | `volume >= 1000000` |
| `==` | 等于 | `signal == "BUY"` |
| `!=` | 不等于 | `status != "closed"` |

```python
price = 178.52
target = 180.00

print(f"当前价: ${price}   目标价: ${target}")
print(f"price > target?  → {price > target}")
print(f"price < target?  → {price < target}")
print(f"price == target? → {price == target}")
```

## 2️⃣ if / elif / else — 交易信号判断

RSI（相对强弱指数）是最常用的超买超卖指标：
- RSI < 30 → **超卖** → 考虑买入 🟢
- RSI > 70 → **超买** → 考虑卖出 🔴
- 其他 → 观望 ⚪

```python
rsi = 25  # ← 试试改成不同的值：25, 50, 75

if rsi < 30:
    signal = "🟢 买入信号（超卖）"
elif rsi > 70:
    signal = "🔴 卖出信号（超买）"
else:
    signal = "⚪ 观望（中性区间）"

print(f"RSI = {rsi}")
print(f"信号: {signal}")

# 可视化 RSI 位置
bar = "█" * (rsi // 2) + "░" * (50 - rsi // 2)
print(f"\n超卖 |{bar}| 超买")
print(f"  0  {'':>{rsi // 2}}▲{'':<{50 - rsi // 2}} 100")
```

## 3️⃣ 多条件组合 — and / or / not

```python
price = 65000
volume = 5_000_000_000   # 50亿（Python中下划线分隔数字，更好读）
ma20 = 63000             # 20日均线

print("=" * 40)
print(f"  价格: ${price:,}")
print(f"  MA20: ${ma20:,}")
print(f"  成交量: {volume/1e9:.1f}B")
print("=" * 40)

# and：所有条件都满足
if price > ma20 and volume > 1_000_000_000:
    print("✅ 满足买入：价格在均线上方 + 成交量放大")
else:
    print("❌ 不满足买入条件")

# or：任一条件满足
if price > 70000 or price < 55000:
    print("⚠️  价格异常波动预警！")
else:
    print("✅ 价格在正常范围内")
```

## 4️⃣ for 循环 — 遍历价格数据

```python
# 模拟一周 BTC 收盘价
days =   ["周一", "周二", "周三", "周四", "周五"]
prices = [64500,  65200,  63800,  66100,  65432]

print("📅 本周 BTC 日报")
print("-" * 35)

for day, price in zip(days, prices):
    bar = "█" * int((price - 63000) / 100)  # 简易柱状图
    print(f"  {day} ${price:,}  {bar}")
```

## 5️⃣ 实战：手算 3 日移动平均线 (SMA)

移动平均线是量化交易最基础的指标：

$$SMA(3) = \frac{P_{t} + P_{t-1} + P_{t-2}}{3}$$

```python
prices = [100, 102, 101, 105, 103, 107, 106, 110, 108, 112]
window = 3

print(f"{'日期':>4} {'价格':>6} {'SMA(3)':>8}  可视化")
print("-" * 45)

sma_values = []
for i in range(len(prices)):
    if i < window - 1:
        print(f"  D{i+1:>2}  {prices[i]:>5}    ---     (数据不足)")
        sma_values.append(None)
    else:
        sma = sum(prices[i-window+1:i+1]) / window
        sma_values.append(sma)
        # 用 ● 和 ○ 表示价格和 SMA
        p_pos = int((prices[i] - 95) / 0.5)
        s_pos = int((sma - 95) / 0.5)
        chart = [' '] * 40
        chart[min(s_pos, 39)] = '○'
        chart[min(p_pos, 39)] = '●'
        print(f"  D{i+1:>2}  {prices[i]:>5}  {sma:>7.2f}  {''.join(chart)}")

print("\n● = 价格  ○ = SMA(3)")
```

## 6️⃣ 统计涨跌天数 + 胜率

```python
daily_prices = [64500, 65200, 63800, 66100, 65432, 67000, 64800, 68200, 67500, 69100]

up_days = 0
down_days = 0
results = []

for i in range(1, len(daily_prices)):
    change = daily_prices[i] - daily_prices[i-1]
    pct = change / daily_prices[i-1] * 100
    
    if change > 0:
        up_days += 1
        emoji = "📈"
    else:
        down_days += 1
        emoji = "📉"
    results.append((i, daily_prices[i], change, pct, emoji))

print("📊 BTC 10日涨跌统计")
print("-" * 45)
for day, price, chg, pct, emoji in results:
    print(f"  Day{day:>2} ${price:>6,}  {emoji} {chg:>+6,} ({pct:>+.2f}%)")

win_rate = up_days / (up_days + down_days) * 100
print(f"\n📈 上涨: {up_days}天  📉 下跌: {down_days}天  胜率: {win_rate:.0f}%")
```

## 7️⃣ while 循环：模拟止损止盈 🎯

设定止损价和止盈价，模拟持仓直到触发其中一个。

```python
import random
random.seed(42)  # 固定随机种子，每次结果一样

entry_price = 100.0
stop_loss = 95.0     # 止损 -5%
take_profit = 110.0  # 止盈 +10%
current = entry_price

print(f"入场: ${entry_price}  止损: ${stop_loss}  止盈: ${take_profit}")
print("=" * 50)

day = 0
history = [current]

while True:
    day += 1
    change = random.uniform(-0.03, 0.03)  # 日波动 ±3%
    current *= (1 + change)
    history.append(current)
    
    # 可视化当前价格位置
    pos = int((current - 90) / 0.5)
    pos = max(0, min(pos, 49))
    bar = list('·' * 50)
    sl_pos = max(0, min(int((stop_loss - 90) / 0.5), 49))
    tp_pos = max(0, min(int((take_profit - 90) / 0.5), 49))
    bar[sl_pos] = '🔴'[0]  
    bar[tp_pos] = '🟢'[0]
    bar[pos] = '▓'
    
    status = ""
    if current <= stop_loss:
        status = " ← 💥 止损！"
    elif current >= take_profit:
        status = " ← 🎉 止盈！"
    
    print(f"Day{day:>2} ${current:>7.2f}{status}")
    
    if current <= stop_loss or current >= take_profit or day >= 30:
        break

pnl = (current - entry_price) / entry_price * 100
print(f"\n最终收益率: {pnl:+.2f}%")
```

## 8️⃣ 列表推导式 — Python 的简洁写法

```python
prices = [100, 102, 101, 105, 103]

# 一行计算每日收益率
returns = [(prices[i] - prices[i-1]) / prices[i-1] * 100 
           for i in range(1, len(prices))]

print(f"价格: {prices}")
print(f"收益率: {[f'{r:+.2f}%' for r in returns]}")

# 筛选上涨的日子
up_returns = [r for r in returns if r > 0]
print(f"上涨日: {[f'{r:+.2f}%' for r in up_returns]}")
```
