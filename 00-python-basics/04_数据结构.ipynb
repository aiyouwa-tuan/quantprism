| 数据结构 | 符号 | 量化用途 |
|---------|------|----------|
| 列表 `list` | `[]` | 价格序列、交易记录 |
| 字典 `dict` | `{}` | 持仓信息、策略参数 |
| 元组 `tuple` | `()` | 不可变的交易记录 |
| 集合 `set` | `{}` | 去重（持仓品种列表）|

## 1️⃣ 列表 — 价格序列

```python
prices = [100.0, 102.5, 99.8, 105.3, 103.1, 108.0, 106.5]

print(f"价格列表: {prices}")
print(f"第1天: {prices[0]}  最后1天: {prices[-1]}")
print(f"前3天: {prices[:3]}")
print(f"后3天: {prices[-3:]}")
print()
print(f"📊 统计")
print(f"  最高: {max(prices)}  最低: {min(prices)}")
print(f"  均价: {sum(prices)/len(prices):.2f}")
print(f"  天数: {len(prices)}")

# 可视化
for i, p in enumerate(prices):
    bar = "█" * int((p - 95) * 2)
    print(f"  Day{i+1} ${p:>6.1f} {bar}")
```

## 2️⃣ 字典 — 持仓管理

```python
# 用字典管理持仓
portfolio = {
    "AAPL": {"shares": 100, "cost": 150.0, "current": 178.5},
    "BTC":  {"shares": 2,   "cost": 60000, "current": 65000},
    "ETH":  {"shares": 10,  "cost": 3000,  "current": 3400},
}

print("┌─────────────────────────────────────────────────┐")
print("│              💼 持仓明细表                        │")
print("├───────┬────────┬──────────┬──────────┬──────────┤")
print("│ 品种  │ 数量   │ 成本     │ 市值     │ 盈亏     │")
print("├───────┼────────┼──────────┼──────────┼──────────┤")

total_cost = 0
total_value = 0

for symbol, info in portfolio.items():
    cost = info["shares"] * info["cost"]
    value = info["shares"] * info["current"]
    pnl = value - cost
    pnl_pct = pnl / cost * 100
    emoji = "📈" if pnl > 0 else "📉"
    
    total_cost += cost
    total_value += value
    
    print(f"│ {symbol:<5} │ {info['shares']:>6} │ ${cost:>8,.0f} │ ${value:>8,.0f} │ {emoji}{pnl_pct:>+6.1f}% │")

total_pnl = (total_value - total_cost) / total_cost * 100
print("├───────┴────────┼──────────┼──────────┼──────────┤")
print(f"│ 合计           │ ${total_cost:>8,.0f} │ ${total_value:>8,.0f} │ {'📈' if total_pnl>0 else '📉'}{total_pnl:>+6.1f}% │")
print("└────────────────┴──────────┴──────────┴──────────┘")
```

## 3️⃣ 集合 — 去重

```python
trades_today = ["AAPL", "BTC", "AAPL", "ETH", "BTC", "AAPL"]
unique = set(trades_today)

print(f"今日交易: {trades_today}")
print(f"去重后: {unique}")
print(f"交易品种数: {len(unique)}")

# 集合运算
watchlist = {"AAPL", "GOOGL", "MSFT", "TSLA"}
holdings = {"AAPL", "BTC", "ETH"}

print(f"\n关注且持有:   {watchlist & holdings}")
print(f"关注但未买:   {watchlist - holdings}")
```

## 4️⃣ 实战：交易日志系统

```python
trade_log = []

def add_trade(symbol, side, price, qty):
    trade_log.append({
        "symbol": symbol,
        "side": side,
        "price": price,
        "qty": qty,
        "value": price * qty
    })

# 记录交易
add_trade("AAPL",  "BUY",  150.0,  100)
add_trade("BTC",   "BUY",  63000,  0.5)
add_trade("AAPL",  "SELL", 178.0,  100)
add_trade("ETH",   "BUY",  3200,   5)

# 打印日志
print("📋 交易日志")
print("─" * 55)
print(f"{'品种':<6} {'方向':<5} {'价格':>10} {'数量':>8} {'金额':>12}")
print("─" * 55)
for t in trade_log:
    side_emoji = "🟢" if t["side"] == "BUY" else "🔴"
    print(f"{t['symbol']:<6} {side_emoji}{t['side']:<4} ${t['price']:>9,.2f} {t['qty']:>8} ${t['value']:>11,.2f}")

buy_total = sum(t["value"] for t in trade_log if t["side"] == "BUY")
sell_total = sum(t["value"] for t in trade_log if t["side"] == "SELL")
print("─" * 55)
print(f"买入总额: ${buy_total:,.2f}  卖出总额: ${sell_total:,.2f}")
```

## 5️⃣ 嵌套结构：K 线数据

```python
# 模拟 K 线数据（和 API 返回格式一样）
klines = [
    {"date": "01-01", "open": 100, "high": 105, "low": 98,  "close": 103, "vol": 1000000},
    {"date": "01-02", "open": 103, "high": 108, "low": 102, "close": 107, "vol": 1200000},
    {"date": "01-03", "open": 107, "high": 110, "low": 104, "close": 105, "vol": 900000},
    {"date": "01-04", "open": 105, "high": 112, "low": 103, "close": 111, "vol": 1500000},
    {"date": "01-05", "open": 111, "high": 115, "low": 109, "close": 113, "vol": 1800000},
]

# 提取收盘价列表（用列表推导式）
closes = [k["close"] for k in klines]
print(f"收盘价: {closes}")

# 找成交量最大的一天
max_vol_day = max(klines, key=lambda k: k["vol"])
print(f"最大成交量: {max_vol_day['date']} ({max_vol_day['vol']:,})")

# 简易 K 线可视化
print("\n📊 简易 K 线")
for k in klines:
    body_low = min(k["open"], k["close"])
    body_high = max(k["open"], k["close"])
    is_up = k["close"] >= k["open"]
    char = "█" if is_up else "░"
    
    line = list(" " * 40)
    for p in range(int(k["low"]) - 95, int(k["high"]) - 95 + 1):
        if 0 <= p < 40:
            if int(body_low) - 95 <= p <= int(body_high) - 95:
                line[p] = char
            else:
                line[p] = "│"
    
    emoji = "🟢" if is_up else "🔴"
    print(f"  {k['date']} {emoji} {''.join(line)}  C:{k['close']}")
```
